### Ipl Score Prediction

```python
import pandas as pd
import pickle
```

```python
# Loading the dataset
df = pd.read_csv('ipl.csv')
```

```python
df.head()
```

```python
df['venue'].unique()
```

```python
consistent_venue = ['M Chinnaswamy Stadium', 'Punjab Cricket Association Stadium, Mohali', 'Feroz Shah Kotla',
                   'Wankhede Stadium', 'Eden Gardens', 'MA Chidambaram Stadium, Chepauk', 'Vidarbha Cricket Association Stadium, Jamtha',
                   'Nehru Stadium', 'Maharashtra Cricket Association Stadium', 'Punjab Cricket Association IS Bindra Stadium, Mohali']
```

### Data Cleaning

```python
# Removing unwanted columns
columns_to_remove = ['mid', 'batsman', 'bowler', 'striker', 'non-striker']
df.drop(labels=columns_to_remove, axis=1, inplace=True)
```

```python
# Keeping only consistent teams
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                    'Mumbai Indians', 'Kings XI Punjab', 'Royal Challengers Bangalore',
                    'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
df = df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
```

```python
df.head()
```

```python
df = df[df['venue'].isin(consistent_venue)]
```

```python
df.head()
```

```python
# Removing the first 5 overs data in every match
df = df[df['overs']>=5.0]
```

```python
# Converting the column 'date' from string into datetime object
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

### Data Preprocessing

```python
# Converting categorical features using OneHotEncoding method
encoded_df = pd.get_dummies(data=df, columns=['venue', 'bat_team', 'bowl_team'])
```

```python
encoded_df.head()
```

```python
encoded_df.columns
```

```python
# Rearranging the columns
encoded_df = encoded_df[['date','venue_Eden Gardens', 'venue_Feroz Shah Kotla',
                         'venue_M Chinnaswamy Stadium', 'venue_MA Chidambaram Stadium, Chepauk',
                         'venue_Maharashtra Cricket Association Stadium',
                         'venue_Punjab Cricket Association IS Bindra Stadium, Mohali',
                         'venue_Punjab Cricket Association Stadium, Mohali',
                         'venue_Wankhede Stadium', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
                         'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
                         'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
                         'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
                         'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
                         'bowl_team_Royal Challengers Bangalore', 'bowl_team_Sunrisers Hyderabad',
                         'overs', 'runs', 'wickets', 'runs_last_5', 'wickets_last_5', 'total']]
```

```python
# Splitting the data into train and test set
X_train = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year <= 2016]
X_test = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year >= 2017]
```

```python
y_train = encoded_df[encoded_df['date'].dt.year <= 2016]['total'].values
y_test = encoded_df[encoded_df['date'].dt.year >= 2017]['total'].values
```

```python
# Removing the 'date' column
X_train.drop(labels='date', axis=True, inplace=True)
X_test.drop(labels='date', axis=True, inplace=True)
```

### Model Buildings

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# Creating a pickle file for the classifier
filename = 'first-innings-score-lr-model.pkl'
pickle.dump(regressor, open(filename, 'wb'))
```
